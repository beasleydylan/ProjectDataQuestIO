# Exploring Hacker News Post Data
Within this code I am exploring the hacker_news.csv which is provided by dataquest.

```python
from csv import reader
file_open = open('hacker_news.csv')
file_read = reader(file_open)
hacker_list = list(file_read)
headers = hacker_list[0]
hn = hacker_list[1:]
for i in hn[:5]:
    print(i)
```

```python
ask_posts, show_posts, other_posts = [], [], []

for i in hn:
    title = i[1]
    if title.lower().startswith('ask hn'):
        ask_posts.append(i)
    if title.lower().startswith('show hn'):
        show_posts.append(i)
    else:
        other_posts.append(i)
print(len(ask_posts), len(show_posts), len(other_posts))
```

```python
print(ask_posts[:5])
print(show_posts[:5])
```

```python
total_ask_comments = 0
for i in ask_posts:
    total_ask_comments += int(i[4])
avg_ask_comments = total_ask_comments/len(ask_posts)
print(avg_ask_comments)

total_show_comments = 0
for i in show_posts:
    total_show_comments += int(i[4])
avg_show_comments = total_show_comments/len(show_posts)
print(avg_show_comments)
```

It seems ask_posts receives more comments per post on average than show posts. Maybe this is due to the fact that ask posts require some sort of community engagement in terms of writing comments rather than, a post about a product would receive. I would like to compare the overral reactions, or likes that show_posts receive to see which receives more community involvement.

```python
import datetime as dt
result_list = []

for i in ask_posts:
    result_list.append([i[6], int(i[4])])

counts_by_hour = {} 
comments_by_hour = {}

for i in result_list:
    time = i[0]
    first = dt.datetime.strptime(time, '%m/%d/%Y %H:%M')
    hour = first.strftime('%H')
    if hour not in counts_by_hour and hour not in comments_by_hour:
        counts_by_hour[hour] = 1
        comments_by_hour[hour] = i[1]
    else:    
        counts_by_hour[hour] += 1
        comments_by_hour[hour] += i[1]
print(counts_by_hour, '\n')
print(comments_by_hour)
```

```python
avg_by_hour = []
for i in counts_by_hour:
    avg_by_hour.append([i, comments_by_hour[i]/counts_by_hour[i]])
print(avg_by_hour)
```

```python
swap_avg_by_hour = []
for i in avg_by_hour:
    swap_avg_by_hour.append([i[1], i[0]])
print(swap_avg_by_hour)
```

I have assorted in the list in descending order. I have found that the max value of average comments would be at 3pm. This would make sense because many people have down time around 3pm. Kids getting out of school, adults day is winding down, some people even take lunch break at this late of hour. So the site probably peaks in usage at this time making more comment available to occur.

```python
sorted_swap = sorted(swap_avg_by_hour, reverse=True)
print('Top 5 Hours for Ask Posts Comments')
for i in sorted_swap[:5]:
    first = dt.datetime.strptime(i[1], '%H')
    time_format = first.strftime('%H:%M')
    print('{0}: {1:.2f} average comments per post.'.format(time_format, i[0]))
```

In this next segment of code I am going to determine the number of points received both both ask and show posts, and which one receives the most points. Then from that, I will gather which one receives the most points and determine at which times the most points would be received.

```python
tot_avg_ask = 0
tot_avg_show = 0

for i in ask_posts:
    tot_avg_ask += int(i[3])
for i in show_posts:
    tot_avg_show += int(i[3])

tot_avg_ask = tot_avg_ask/len(ask_posts)
tot_avg_show = tot_avg_show/len(show_posts)

print(tot_avg_ask, tot_avg_show)
```

It seems I was correct in my assumption in that show_posts would receive more like on average than ask_posts would receive. From here I will determine for show_posts at what times per hour the most likes would be received.

```python
point_per_hour = {}
count_per_hour = {}
print(len(show_posts))
for i in show_posts:
    dat = dt.datetime.strptime(i[6], '%m/%d/%Y %H:%M')
    here = dat.strftime('%H')
#    print(here)
    if here in point_per_hour and here in count_per_hour:
        count_per_hour[here] += 1
        point_per_hour[here] += int(i[4])
#        print(i[4])
        continue
    count_per_hour[here] = 1
    point_per_hour[here] = int(i[4])
print(point_per_hour)
print(count_per_hour)
```

Now that I have the values for the number of posts per hour, and the total number of like per hour. I will now formulate a data structure that will have the average number of likes per hour.

```python
tot_dict = {}
lis_avg = []
for i in point_per_hour:
    lis_avg.append([(point_per_hour[i]/count_per_hour[i]), i])
fin_lis = sorted(lis_avg, reverse=True)

print('Top 5 Average Likes Per Show Posts')
for i in fin_lis[:5]:
    thanks = dt.datetime.strptime(i[1], '%H')
    thanks = thanks.strftime('%H:%M')
    print('{0}: {1:.2f} average points per post per hour.'.format(thanks, i[0]))
```

From this data we can conclude that likes per hour is most accumulated when the post is at 6pm.

```python
com_point = []
print(headers)
for i in other_posts:
        com_point.append([i[3], i[4], i[6]])
print(com_point)
```

I made a list holding the times, number or points, and number of comments each post has in other posts. From here I will separate the number of posts per hour and the individual points and comments to individual dictionaries

```python
com_dict = {}
point_dict = {}
count_dict = {}

for i in com_point:
    time = dt.datetime.strptime(i[2], '%m/%d/%Y %H:%M')
    time = time.strftime('%H')
    # only need to compare one dict because all have same keys
    if time in com_dict:
        com_dict[time] += int(i[1])
        point_dict[time] += int(i[0])
        count_dict[time] += 1
        continue
    com_dict[time] = int(i[1])
    point_dict[time] = int(i[0])
    count_dict[time] = 1

print(com_dict, '\n')
print(point_dict, '\n')
print(count_dict, '\n')
```

I have assorted each dictionary for points accumlated per hour, comments accumlated per hour, and the total number of posts per hour. From here I will obtain the average number of points and comments per hour. Then compare them to show_posts and ask_posts in accordance to which have the most comments and points.

```python
avg_com_other = []
avg_point_other = []
for i in com_dict:
    avg_com_other.append([com_dict[i]/count_dict[i], i])
    avg_point_other.append([point_dict[i]/count_dict[i], i])
final_com = sorted(avg_com_other, reverse=True)
final_point = sorted(avg_point_other, reverse=True)

print('Top 5 Comments per Hour for Other Posts:')
for i in final_com[:5]:
    here = dt.datetime.strptime(i[1], '%H')
    here = here.strftime('%H:%M')
    print('{0}: {1:.2f} average comments per post per hour.'.format(here, i[0]))
print('\nTop 5 Comments per Hour for Other Posts:')
for i in final_point[:5]:
    here = dt.datetime.strptime(i[1], '%H')
    here = here.strftime('%H:%M')
    print('{0}: {1:.2f} average points per post per hour.'.format(here, i[0]))
print('\n', )
print(len(other_posts), len(ask_posts), len(show_posts))
```

It seems Other Posts overall has more points on average compared to ask and show posts, but less comments than ask posts on average.
Show posts had more overall points compared to ask, so I will only use that data to compare: 18:00: 15.77 average points per post per hour.
Show posts only had 15.77 average points while other posts had 59.28 average points per hour. This may be due to the fact that other posts has 18,938 posts and show only has 1,162.
Ask posts had 38.59 comements per hour while Other Posts had 30.43. This is probably due to the fact that ask posts require more comments within the community rather than other posts.
